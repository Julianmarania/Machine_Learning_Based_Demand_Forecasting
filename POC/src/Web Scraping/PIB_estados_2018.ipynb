```python
time.sleep(10)

rows = driver.find_elements(By.CSS_SELECTOR, 'table#DataTbl tbody tr')

datos = []

# Función para limpiar y convertir los valores a float, manejando los puntos de miles
def limpiar_valor(valor):
    if valor:
        valor = valor.strip()
        # Eliminar las comas (separadores de miles)
        valor = valor.replace(',', '')
        try:
            float_valor = float(valor)
            print(f"Conversión a float exitosa (solo eliminar comas): {float_valor} (original: '{valor}')")
            return float_valor
        except ValueError:
            print(f"Error al convertir a float (solo eliminar comas): '{valor}' (original: '{valor}')")
            return 0
    return 0


# Iterar sobre cada fila
for row in rows:
    cols = row.find_elements(By.TAG_NAME, 'td')

    if len(cols) >= 4:
        cp = cols[0].text.strip()
        estado = cols[1].text.strip()

        q1 = limpiar_valor(cols[2].text)
        q2 = limpiar_valor(cols[3].text)

        q1 *= 1_000_000_000
        q2 *= 1_000_000_000

        # Añadir los datos procesados a la lista
        datos.append([cp, estado, int(q1), int(q2)])

# Crear DataFrame de Pandas con los datos extraídos
pib_estados = pd.DataFrame(datos, columns=["cp", "estado", "Q1", "Q2"])

# Mostrar el DataFrame
print(pib_estados)

# Cerrar el navegador
driver.quit()
```

```python
pib_estados['january'] = pib_estados['Q1'] / 3

pib_estados['february'] = pib_estados['Q1'] / 3

pib_estados['march'] = pib_estados['Q1'] / 3

pib_estados['april'] = pib_estados['Q2'] / 3

pib_estados['may'] = pib_estados['Q2'] / 3

pib_estados = pib_estados.drop(columns = ["Q1", "Q2"])

pib_estados
```
