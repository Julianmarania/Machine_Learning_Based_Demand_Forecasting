```python
# Esperar a que la tabla cargue
time.sleep(10)

# Extraer filas
rows = driver.find_elements(By.CSS_SELECTOR, 'table#DataTbl tbody tr')

datos = []
for row in rows:
    cols = row.find_elements(By.TAG_NAME, 'td')
    if len(cols) >= 4:
        cp = cols[0].text.strip()  
        estado = cols[1].text.strip().replace(' *', '').replace(' ', '')
        q1 = cols[2].text.strip().replace('.', '') 
        q1 = int(float(q1.replace(',', '.')) * 1_000_000_000)
        q2 = cols[3].text.strip().replace('.', '') 
        q2 = int(float(q2.replace(',', '.')) * 1_000_000_000) 
        datos.append([cp, estado, q1, q2])

# Crear DataFrame
capital_personal_income = pd.DataFrame(datos, columns=["cp", "estado", "Q1", "Q2"])
print(capital_personal_income)

# Cerrar navegador
driver.quit()
```

```python
capital_personal_income['january'] = capital_personal_income['Q1'] / 3

capital_personal_income['february'] = capital_personal_income['Q1'] / 3

capital_personal_income['march'] = capital_personal_income['Q1'] / 3

capital_personal_income['april'] = capital_personal_income['Q2'] / 3

capital_personal_income['may'] = capital_personal_income['Q2'] / 3

capital_personal_income = capital_personal_income.drop(columns = ["Q1", "Q2"])

capital_personal_income
```

```python
## Guardar en CSV
# Path
capital_personal_income_path = os.path.join(current_dir, '../../data/raw/macrodata')
# Crear un directorio si no existe
os.makedirs(capital_personal_income_path, exist_ok=True)
# Archivo CSV
capital_personal_income_file = os.path.join(capital_personal_income_path, "riqueza_per_capita.csv")
# Guardar dataframe en CSV
capital_personal_income.to_csv(capital_personal_income_file, index=False)
```
