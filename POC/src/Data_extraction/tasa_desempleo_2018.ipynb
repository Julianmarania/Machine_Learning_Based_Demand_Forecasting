```python
import pandas as pd
import os

# Current dir
current_dir = os.getcwd()
```

```python
url_tasa_desempleo = os.path.join(current_dir, '../../data/raw/extractions/tasa_desempleo_2018.csv')

tasa_desempleo = pd.read_csv(url_tasa_desempleo)

tasa_desempleo = tasa_desempleo.drop(columns = ["GeoID", "Series ID"])

tasa_desempleo = tasa_desempleo[tasa_desempleo['Month'].isin(['January', 'February', 'March', 'April', 'May'])]

tasa_desempleo
```

```python
## Guardar en CSV
# Path
tasa_desempleo_path = os.path.join(current_dir, '../../data/raw/macrodata')
# Crear un directorio si no existe
os.makedirs(tasa_desempleo_path, exist_ok=True)
# Archivo CSV
tasa_desempleo_file = os.path.join(tasa_desempleo_path, "tasa_desempleo_por_estados.csv")
# Guardar dataframe en CSV
tasa_desempleo.to_csv(tasa_desempleo_file, index=False)
```
