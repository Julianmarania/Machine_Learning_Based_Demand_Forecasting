```python
import pandas as pd
import os

# Current dir
current_dir = os.getcwd()
```

```python
url_nivel_estudios = os.path.join(current_dir, '../../data/raw/extractions/studies-Data.csv')

nivel_estudios = pd.read_csv(url_nivel_estudios)

#Queremos que el encabezado sea la fila 0

# Usar la fila 0 como nombres de columnas
nivel_estudios.columns = nivel_estudios.iloc[0]  # Establece los valores de la fila 0 como encabezados

# Eliminar la fila 0 del DataFrame
nivel_estudios = nivel_estudios.drop(0).reset_index(drop=True)

nivel_estudios = nivel_estudios.drop(columns = ["Geography"])

nivel_estudios = nivel_estudios.dropna(axis=1, how='all')

def clean_column_name(col):
    # Convertir a string por si acaso
    col = str(col)
    # Eliminar espacios iniciales y finales
    col = col.strip()
    # Eliminar comillas iniciales y finales
    col = col.strip("'")
    # Eliminar prefijos como 'Estimate!!Total!!' o 'Estimate!!'
    col = col.replace('Estimate!!Total!!', '').replace('Estimate!!', '')
    # Eliminar texto redundante como 'Population 18 to 24 years' o 'Population 25 years and over'
    col = col.replace('Population 18 to 24 years!!', '').replace('Population 25 years and over!!', '')
    # Reemplazar espacios por guiones bajos y eliminar caracteres especiales
    col = col.replace(' ', '_').replace('!!', '_').replace('__', '_')
    # Convertir a minúsculas para consistencia (opcional)
    col = col.lower()
    # Eliminar guiones bajos al inicio o final
    col = col.strip('_')
    return col

nivel_estudios.columns = [clean_column_name(col) for col in nivel_estudios.columns]

nivel_estudios
```

```python
## Guardar en CSV
# Path
nivel_estudios_path = os.path.join(current_dir, '../../data/raw/macrodata')

# Crear un directorio si no existe
os.makedirs(nivel_estudios_path, exist_ok=True)  # Cambié 'nivel_estudios' por 'nivel_estudios_path'

# Archivo CSV
nivel_estudios_file = os.path.join(nivel_estudios_path, "nivel_estudios_por_estados.csv")

# Guardar dataframe en CSV
nivel_estudios.to_csv(nivel_estudios_file, index=False)
```
